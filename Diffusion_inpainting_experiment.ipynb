```python
import os
import hashlib
import shutil

# Base directory where files are stored
weights_dir = "weights"

# Dictionary to store file hashes
file_hashes = {}

def get_file_hash(file_path):
    """Calculate SHA-256 hash of a file to detect duplicates."""
    hasher = hashlib.sha256()
    with open(file_path, "rb") as f:
        while chunk := f.read(8192):
            hasher.update(chunk)
    return hasher.hexdigest()

def remove_duplicates(directory):
    """Scan through all files and remove duplicates, keeping only one copy."""
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)

            # Get file hash
            file_hash = get_file_hash(file_path)

            if file_hash in file_hashes:
                print(f"🗑️ Deleting duplicate: {file_path}")
                os.remove(file_path)  # Delete duplicate
            else:
                file_hashes[file_hash] = file_path  # Store unique file hash

    print("\n✅ Duplicate cleanup completed!")

# Run the function
remove_duplicates(weights_dir)
```

```python
import os
import shutil

# Base directory containing weights
base_dir = "weights"

# Cache directories to completely delete
cache_dirs = [
    os.path.join(base_dir, ".cache/huggingface/download"),
    os.path.join(base_dir, "stable-diffusion-v1-5/.cache/huggingface/download"),
    os.path.join(base_dir, "diffuEraser/unet_main/.cache/huggingface/download"),
    os.path.join(base_dir, "diffuEraser/brushnet/.cache/huggingface/download"),
    os.path.join(base_dir, "sd-vae-ft-mse/.cache/huggingface/download"),
]

# Step 1: Delete the entire cache directories
for cache_dir in cache_dirs:
    if os.path.exists(cache_dir):
        print(f"🗑 Deleting entire cache directory: {cache_dir}")
        shutil.rmtree(cache_dir)

# Step 2: Remove any empty `.cache/huggingface/` directories left behind
cache_main = os.path.join(base_dir, ".cache/huggingface")
if os.path.exists(cache_main) and not os.listdir(cache_main):
    print(f"📁 Removing empty directory: {cache_main}")
    shutil.rmtree(cache_main)

print("\n✅ Cleanup complete! All unnecessary cache files have been deleted.")
```

```python
import os
import shutil

# Base directory
base_dir = "weights"

# Ensure PCM_Weights exists
pcm_weights_dir = os.path.join(base_dir, "PCM_Weights")
os.makedirs(pcm_weights_dir, exist_ok=True)

# Move sd15 into PCM_Weights
sd15_path = os.path.join(base_dir, "sd15")
new_sd15_path = os.path.join(pcm_weights_dir, "sd15")
if os.path.exists(sd15_path):
    print(f"🚚 Moving {sd15_path} → {new_sd15_path}")
    shutil.move(sd15_path, new_sd15_path)

# Cache directories to delete
cache_dirs = [
    os.path.join(base_dir, ".cache"),
    os.path.join(base_dir, "stable-diffusion-v1-5/.cache"),
    os.path.join(base_dir, "diffuEraser/unet_main/.cache"),
    os.path.join(base_dir, "diffuEraser/brushnet/.cache"),
    os.path.join(base_dir, "sd-vae-ft-mse/.cache"),
]

# Delete unnecessary cache folders
for cache_dir in cache_dirs:
    if os.path.exists(cache_dir):
        print(f"🗑 Deleting {cache_dir}")
        shutil.rmtree(cache_dir)

# Create a README file
readme_path = os.path.join(base_dir, "README.md")
if not os.path.exists(readme_path):
    print(f"📄 Creating {readme_path}")
    with open(readme_path, "w") as f:
        f.write("# Weights Directory Structure\n\nThis directory contains all necessary models and weights.")

print("\n✅ Cleanup & organization complete! Your folder structure is now correct.")
```
